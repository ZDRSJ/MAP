```python
import pandas as pd
import folium
import webbrowser
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv('/content/drive/MyDrive/부동산_실거래가_예측/map구현/부동산_위치정보_new.csv')
df
```

```python
map = folium.Map(location = [37.8797296,126.4964884], zoom_start = 8)
marker_cluster = MarkerCluster().add_to(map)
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
for index,a in df.iterrows():
  # print(a['위도'])
  popup = f"<pre>{str(a['아파트이름'])}<br>전용면적 : {str(a['전용면적'])}<br>평균 거래금액 : {str(round(a['평균 거래금액']))}</pre>"
  folium.Marker(location = [a['위도'],a['경도']],popup = popup).add_to(marker_cluster)
```
